```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])

y = np.array([1.2, 1.8, 2.6, 3.2, 3.8])
```

```python
x_mean = np.mean(x)

y_mean = np.mean(y)

print(x_mean)
print(y_mean)
```

```python
sum1 = 0
sum2 = 0
for i in range(0, len(x)):
    sum1 = (x[i] - x_mean) * (y[i] - y_mean)
    sum2 = (x[i] - x_mean)**2

m = sum1/sum2

b = y_mean - m*x_mean

print(f"Intercept: {m}, Bias: {b}")
```

```python

def model(week):
    return m*week + b


print(model(2))
```

```python
newX = np.arange(0, 12, 1)
```

```python
plt.plot(x, y)
plt.plot(newX, model(newX), c='r')
plt.show()
```
